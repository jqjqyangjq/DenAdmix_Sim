```python
import pandas as pd
snp = pd.read_csv("/mnt/diversity/jiaqi/hmm/hmm_extend_sim_compare_demo/sim/chrom_1.snps.tsv.gz", sep='\t')
snp['afr_total'] = 0
for i in range(0,200):
    snp['afr_total'] = snp['afr_total'] + snp['afr_'+str(i)]
snp = snp[snp['afr_total'] <= 0 ]
archaic = snp.copy()
archaic['NEA'] = snp['nea_2'] + snp['nea_3']
archaic['DEN'] = snp['den3_0'] + snp['den3_1']
archaic['DEN_N'] = snp['Intro_den_N_0'] + snp['Intro_den_N_1']
archaic['DEN_S'] = snp['Intro_den_S_0'] + snp['Intro_den_S_1']
archaic['NEA_out'] = snp['nea_out_0'] + snp['nea_out_1']
```

```python
print(len(archaic[(archaic['NEA'] > 0) & (archaic['DEN'] > 0)]))
print(len(archaic[(archaic['NEA'] > 0) & (archaic['DEN'] == 0)]))
print(len(archaic[(archaic['NEA'] == 0) & (archaic['DEN'] > 0)]))
```
